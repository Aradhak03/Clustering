```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
```

```python
df.head()
```

```python
from pycaret.clustering import *

def modelfunction(data, clustering_algo='kmeans', transformation=False, normalize=False, pca=False,
                  normalize_method='zscore', transformation_method='yeo-johnson', pca_method='linear', verbose=False):

    setup(data=data, transformation=transformation, normalize=normalize, pca=pca,
          normalize_method=normalize_method,
          transformation_method=transformation_method,
          pca_method=pca_method, verbose=verbose)

    print("For Cluster = 3")
    x = create_model(clustering_algo, num_clusters=3)
    metrics1 = pull()

    print("For Cluster = 4")
    x = create_model(clustering_algo, num_clusters=4)
    metrics2 = pull()

    print("For Cluster = 5")
    x = create_model(clustering_algo, num_clusters=5)
    metrics3 = pull()

    newtable = pd.concat([metrics1, metrics2, metrics3], axis=0)
    newtable = newtable.T
    columnname = ['c=3', 'c=4', 'c=5']
    newtable.columns = columnname

    print(newtable)

    return newtable
```

```python
a = modelfunction(df)
print(a)

b = modelfunction(df, clustering_algo='hclust', normalize=True)
print(b)
```

```python
newtable = pd.concat([a, b], axis=1, keys=['No Data Preprocessing', 'Normalize'])
newtable
```

```python
cluster_algo = ['kmeans', 'meanshift', 'hclust']

list_results = []

for i in range(len(cluster_algo)):
    keys = ['With No Data PreProcessing', 'Normalization', 'Transformation', 'PCA', 'Normalize+Transform',
            'Normalize+Transform+PCA']

    a = modelfunction(df, clustering_algo=cluster_algo[i], normalize=True)
    b = modelfunction(df, clustering_algo=cluster_algo[i])
    c = modelfunction(df, clustering_algo=cluster_algo[i], transformation=True)
    d = modelfunction(df, clustering_algo=cluster_algo[i], pca=True)
    e = modelfunction(df, clustering_algo=cluster_algo[i], normalize=True, transformation=True)
    f = modelfunction(df, clustering_algo=cluster_algo[i], normalize=True, transformation=True, pca=True)

    newtable = pd.concat([a, b, c, d, e, f], axis=1, keys=keys)
    list_results.append(newtable)
```

```python
Cluster_names = ['Kmeans Clustering', 'Kmeans Shift Clustering', 'Heirarchial Clustering']

with pd.ExcelWriter('result.xlsx') as writer:
    for i, df_result in enumerate(list_results):
        print(f"Clustering Algo {i + 1}: {Cluster_names[i]}")
        print()
        df_result.to_excel(writer, sheet_name=Cluster_names[i])

print("All Model Result are successfully imported in Excel File with result of each Clustering Algo on Separate Sheet")
```
